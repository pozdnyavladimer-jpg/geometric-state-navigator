```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
from math import pi

np.random.seed(42)

# =========================================================
# 1. CONFIG
# =========================================================
HEX_KEYS = [
    "red_mass",       # pressure / unresolved tension
    "orange_flow",    # motion / adaptability
    "yellow_struct",  # structure / planning integrity
    "green_balance",  # internal consistency
    "blue_law",       # constraint compliance
    "violet_future",  # transition potential
]

COLORS = {
    "red_mass": "#ff5a5a",
    "orange_flow": "#ffb347",
    "yellow_struct": "#ffe44d",
    "green_balance": "#5cff8a",
    "blue_law": "#66b3ff",
    "violet_future": "#d46bff",
}

TARGET = {
    "red_mass": 0.06,
    "orange_flow": 0.22,
    "yellow_struct": 0.22,
    "green_balance": 0.22,
    "blue_law": 0.14,
    "violet_future": 0.14,
}

GRAPH = {
    "A": ["B", "C"],
    "B": ["A", "C", "D"],
    "C": ["A", "B", "E"],
    "D": ["B", "E", "F"],
    "E": ["C", "D", "F"],
    "F": ["D", "E"],
}

POS = {
    "A": (-1.4,  0.8),
    "B": (-0.5,  1.2),
    "C": (-0.5,  0.2),
    "D": ( 0.7,  1.0),
    "E": ( 0.7,  0.0),
    "F": ( 1.5,  0.5),
}

CANDIDATES = {
    "strict_answer": {
        "red_mass": -0.04,
        "orange_flow": -0.01,
        "yellow_struct": +0.03,
        "green_balance": +0.01,
        "blue_law": +0.05,
        "violet_future": -0.01,
    },
    "creative_answer": {
        "red_mass": +0.01,
        "orange_flow": +0.05,
        "yellow_struct": +0.00,
        "green_balance": +0.02,
        "blue_law": -0.02,
        "violet_future": +0.04,
    },
    "balanced_answer": {
        "red_mass": -0.03,
        "orange_flow": +0.03,
        "yellow_struct": +0.03,
        "green_balance": +0.04,
        "blue_law": +0.02,
        "violet_future": +0.01,
    },
    "uncertain_answer": {
        "red_mass": +0.03,
        "orange_flow": -0.01,
        "yellow_struct": -0.01,
        "green_balance": -0.02,
        "blue_law": -0.01,
        "violet_future": +0.02,
    },
}

# =========================================================
# 2. HELPERS
# =========================================================
def normalize(v):
    total = sum(max(v[k], 0.0) for k in HEX_KEYS)
    if total <= 1e-9:
        return {k: 1.0 / len(HEX_KEYS) for k in HEX_KEYS}
    return {k: max(v[k], 0.0) / total for k in HEX_KEYS}

def random_state():
    v = {
        "red_mass": np.random.uniform(0.18, 0.34),
        "orange_flow": np.random.uniform(0.08, 0.18),
        "yellow_struct": np.random.uniform(0.10, 0.20),
        "green_balance": np.random.uniform(0.10, 0.20),
        "blue_law": np.random.uniform(0.08, 0.16),
        "violet_future": np.random.uniform(0.05, 0.12),
    }
    return normalize(v)

def avg_state(field):
    return {
        k: float(np.mean([field[n][k] for n in field]))
        for k in HEX_KEYS
    }

def alignment_distance(v):
    g = v["green_balance"]
    y = v["yellow_struct"]
    o = v["orange_flow"]
    return abs(g - y) + abs(y - o) + abs(g - o)

def target_distance(v):
    return sum(abs(v[k] - TARGET[k]) for k in HEX_KEYS)

def compute_metrics(field):
    avg = avg_state(field)
    shadow = avg["red_mass"]
    coherence = 1.0 - min(1.0, alignment_distance(avg) * 2.0)
    target_fit = 1.0 - min(1.0, target_distance(avg) * 0.9)
    vitality = (
        avg["orange_flow"]
        + avg["yellow_struct"]
        + avg["green_balance"]
        + 0.5 * avg["violet_future"]
    )
    return avg, shadow, coherence, target_fit, vitality

def apply_delta(v, delta):
    out = dict(v)
    for k in HEX_KEYS:
        out[k] += delta.get(k, 0.0)
    return normalize(out)

# =========================================================
# 3. GRAPH MESSAGE PASSING
# =========================================================
def graph_mix(field, alpha=0.25):
    new_field = {}
    for node, v in field.items():
        neigh = GRAPH[node]
        if not neigh:
            new_field[node] = dict(v)
            continue

        neigh_avg = {
            k: float(np.mean([field[n][k] for n in neigh]))
            for k in HEX_KEYS
        }
        mixed = {
            k: (1 - alpha) * v[k] + alpha * neigh_avg[k]
            for k in HEX_KEYS
        }
        new_field[node] = normalize(mixed)
    return new_field

# =========================================================
# 4. GEOMETRY / GSL REGULATOR
# =========================================================
def geometry_tighten(field, k_coh=0.18, k_target=0.10, k_shadow=0.12):
    """
    Pulls states toward:
    - lower red_mass
    - tighter G/Y/O coherence
    - mild target attraction
    """
    new_field = {}
    for node, v in field.items():
        out = dict(v)

        # 1) reduce pressure
        out["red_mass"] *= (1.0 - k_shadow)

        # 2) tighten G/Y/O locally
        local_mean = (
            out["green_balance"]
            + out["yellow_struct"]
            + out["orange_flow"]
        ) / 3.0

        out["green_balance"] += (local_mean - out["green_balance"]) * k_coh
        out["yellow_struct"] += (local_mean - out["yellow_struct"]) * k_coh
        out["orange_flow"]   += (local_mean - out["orange_flow"]) * k_coh

        # 3) mild attraction to target
        for k in HEX_KEYS:
            out[k] += (TARGET[k] - out[k]) * k_target

        new_field[node] = normalize(out)

    return new_field

def gate_step(field, candidate_name, focus_node="C"):
    """
    Apply candidate to one focus node, then run:
    graph propagation -> geometry tightening -> graph propagation
    """
    delta = CANDIDATES[candidate_name]

    # local injection
    stage1 = {n: dict(v) for n, v in field.items()}
    stage1[focus_node] = apply_delta(stage1[focus_node], delta)

    # spread through graph
    stage2 = graph_mix(stage1, alpha=0.30)

    # regulate through geometry
    stage3 = geometry_tighten(stage2, k_coh=0.22, k_target=0.10, k_shadow=0.15)

    # another graph pass
    stage4 = graph_mix(stage3, alpha=0.18)

    return stage4

# =========================================================
# 5. VISUALS
# =========================================================
def draw_graph_field(ax, field, title=""):
    ax.set_facecolor("black")

    # edges
    drawn = set()
    for a, neighs in GRAPH.items():
        for b in neighs:
            key = tuple(sorted((a, b)))
            if key in drawn:
                continue
            drawn.add(key)
            x1, y1 = POS[a]
            x2, y2 = POS[b]
            ax.plot([x1, x2], [y1, y2], color="gray", alpha=0.35, linewidth=1.3)

    # nodes
    for n, v in field.items():
        x, y = POS[n]
        color = (
            v["red_mass"],
            (v["yellow_struct"] + v["orange_flow"]) / 2.0,
            v["green_balance"]
        )
        circ = Circle((x, y), 0.18, color=color, alpha=0.95, ec="white", lw=1.2)
        ax.add_patch(circ)
        ax.text(x, y, n, color="white", ha="center", va="center", fontsize=11, fontweight="bold")

    avg, shadow, coherence, target_fit, vitality = compute_metrics(field)
    subtitle = (
        f"shadow={shadow:.3f} | coherence={coherence:.3f} | "
        f"target_fit={target_fit:.3f} | vitality={vitality:.3f}"
    )
    ax.set_title(f"{title}\n{subtitle}", color="white", fontsize=12)
    ax.set_xlim(-2.0, 2.0)
    ax.set_ylim(-0.6, 1.8)
    ax.axis("off")

def draw_radar(ax, avg, title="Average 6D State"):
    labels = HEX_KEYS
    values = [avg[k] for k in labels]
    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()

    values += values[:1]
    angles += angles[:1]

    ax.set_facecolor("black")
    ax.plot(angles, values, linewidth=2)
    ax.fill(angles, values, alpha=0.25)

    for angle, k in zip(angles[:-1], labels):
        ax.scatter([angle], [avg[k]], s=70, color=COLORS[k], zorder=3)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(
        ["red", "orange", "yellow", "green", "blue", "violet"],
        color="white",
        fontsize=9
    )
    ax.set_yticklabels([])
    ax.set_title(title, color="white", fontsize=12)

# =========================================================
# 6. RUN MVP
# =========================================================
base_field = {node: random_state() for node in GRAPH}
base_avg, base_shadow, base_coh, base_fit, base_vit = compute_metrics(base_field)

results = []
candidate_fields = {}

for cand in CANDIDATES:
    out_field = gate_step(base_field, cand, focus_node="C")
    candidate_fields[cand] = out_field
    avg, shadow, coherence, target_fit, vitality = compute_metrics(out_field)

    # GSL score: prefer lower shadow, higher coherence, target_fit, vitality
    gsl_score = (
        coherence * 0.40
        + target_fit * 0.30
        + min(1.0, vitality) * 0.20
        + (1.0 - shadow) * 0.10
    )

    results.append({
        "candidate": cand,
        "shadow": shadow,
        "coherence": coherence,
        "target_fit": target_fit,
        "vitality": vitality,
        "gsl_score": gsl_score
    })

results = sorted(results, key=lambda x: x["gsl_score"], reverse=True)
best = results[0]["candidate"]
best_field = candidate_fields[best]
best_avg, *_ = compute_metrics(best_field)

# =========================================================
# 7. DISPLAY
# =========================================================
fig = plt.figure(figsize=(15, 8), facecolor="black")
gs = fig.add_gridspec(2, 3, width_ratios=[1.1, 1.1, 1.2], height_ratios=[1, 1], wspace=0.25, hspace=0.25)

ax1 = fig.add_subplot(gs[:, 0])
draw_graph_field(ax1, base_field, title="BASE GRAPH FIELD")

ax2 = fig.add_subplot(gs[:, 1])
draw_graph_field(ax2, best_field, title=f"BEST AFTER GSL: {best}")

ax3 = fig.add_subplot(gs[0, 2], polar=True)
draw_radar(ax3, base_avg, title="Base Average 6D State")

ax4 = fig.add_subplot(gs[1, 2])
ax4.set_facecolor("black")
ax4.axis("off")

lines = ["GSL Candidate Ranking", ""]
for i, row in enumerate(results, start=1):
    lines.append(
        f"{i}. {row['candidate']}\n"
        f"   shadow={row['shadow']:.3f} | "
        f"coh={row['coherence']:.3f} | "
        f"fit={row['target_fit']:.3f} | "
        f"vit={row['vitality']:.3f} | "
        f"score={row['gsl_score']:.3f}"
    )

ax4.text(
    0.02, 0.98,
    "\n\n".join(lines),
    va="top", ha="left",
    color="white",
    fontsize=11,
    family="monospace"
)

plt.show()

print("=== BASE METRICS ===")
print(f"shadow={base_shadow:.3f} | coherence={base_coh:.3f} | target_fit={base_fit:.3f} | vitality={base_vit:.3f}")
print()
print("=== BEST CANDIDATE ===")
print(best)
print()
print("=== RANKING ===")
for row in results:
    print(
        f"{row['candidate']:16s} | "
        f"shadow={row['shadow']:.3f} | "
        f"coherence={row['coherence']:.3f} | "
        f"target_fit={row['target_fit']:.3f} | "
        f"vitality={row['vitality']:.3f} | "
        f"gsl_score={row['gsl_score']:.3f}"
    )
```

```python

# =========================================================
# GSL v2 — TEXT CANDIDATES OVER GRAPH+GEOMETRY
# вставити НИЖЧЕ попередньої клітини
# =========================================================

import re
import matplotlib.pyplot as plt

TEXT_CANDIDATES = {
    "answer_yes": "Yes, proceed with the current plan. The structure is good enough.",
    "answer_strict": "Need more evidence before proceeding. Validate constraints and reduce uncertainty.",
    "answer_creative": "Try an alternative route. There may be a creative way around the current limitation.",
    "answer_uncertain": "I am not sure. The situation is ambiguous and may require more context.",
    "answer_balanced": "Proceed carefully, but keep flexibility. Preserve structure while adapting if needed."
}

def encode_text_to_6d(text: str):
    """
    Простий rule-based encoder:
    текст -> 6D state delta
    Це MVP, не NLP-модель.
    """
    t = text.lower()

    delta = {
        "red_mass": 0.0,
        "orange_flow": 0.0,
        "yellow_struct": 0.0,
        "green_balance": 0.0,
        "blue_law": 0.0,
        "violet_future": 0.0,
    }

    # pressure / uncertainty
    if any(w in t for w in ["not sure", "uncertain", "ambiguous", "unknown", "more context"]):
        delta["red_mass"] += 0.05
        delta["green_balance"] -= 0.02

    if any(w in t for w in ["need more evidence", "validate", "constraints", "reduce uncertainty"]):
        delta["red_mass"] -= 0.03
        delta["blue_law"] += 0.05
        delta["yellow_struct"] += 0.03
        delta["orange_flow"] -= 0.01

    # flow / exploration
    if any(w in t for w in ["alternative", "creative", "adapt", "around", "route", "flexibility"]):
        delta["orange_flow"] += 0.05
        delta["violet_future"] += 0.03
        delta["green_balance"] += 0.02
        delta["blue_law"] -= 0.01

    # structure / plan
    if any(w in t for w in ["plan", "structure", "proceed", "current plan"]):
        delta["yellow_struct"] += 0.03
        delta["green_balance"] += 0.02

    # balance language
    if any(w in t for w in ["carefully", "preserve", "while adapting", "but keep flexibility"]):
        delta["green_balance"] += 0.04
        delta["yellow_struct"] += 0.02
        delta["orange_flow"] += 0.02
        delta["red_mass"] -= 0.02

    # decisive language
    if any(w in t for w in ["yes", "proceed"]):
        delta["red_mass"] -= 0.01
        delta["yellow_struct"] += 0.02

    return delta

def gate_step_text(field, text, focus_node="C"):
    delta = encode_text_to_6d(text)

    stage1 = {n: dict(v) for n, v in field.items()}
    stage1[focus_node] = apply_delta(stage1[focus_node], delta)

    stage2 = graph_mix(stage1, alpha=0.30)
    stage3 = geometry_tighten(stage2, k_coh=0.22, k_target=0.10, k_shadow=0.15)
    stage4 = graph_mix(stage3, alpha=0.18)

    return stage4, delta

# ---------------------------------------------------------
# RUN TEXT-BASED CANDIDATES
# ---------------------------------------------------------
text_results = []
text_fields = {}
text_deltas = {}

for name, text in TEXT_CANDIDATES.items():
    out_field, delta = gate_step_text(base_field, text, focus_node="C")
    text_fields[name] = out_field
    text_deltas[name] = delta

    avg, shadow, coherence, target_fit, vitality = compute_metrics(out_field)

    gsl_score = (
        coherence * 0.40
        + target_fit * 0.30
        + min(1.0, vitality) * 0.20
        + (1.0 - shadow) * 0.10
    )

    text_results.append({
        "candidate": name,
        "text": text,
        "shadow": shadow,
        "coherence": coherence,
        "target_fit": target_fit,
        "vitality": vitality,
        "gsl_score": gsl_score
    })

text_results = sorted(text_results, key=lambda x: x["gsl_score"], reverse=True)
best_text = text_results[0]["candidate"]
best_text_field = text_fields[best_text]
best_text_avg, *_ = compute_metrics(best_text_field)

# ---------------------------------------------------------
# VISUALIZE
# ---------------------------------------------------------
fig = plt.figure(figsize=(16, 9), facecolor="black")
gs = fig.add_gridspec(2, 3, width_ratios=[1.0, 1.0, 1.35], height_ratios=[1, 1], wspace=0.25, hspace=0.25)

ax1 = fig.add_subplot(gs[:, 0])
draw_graph_field(ax1, base_field, title="BASE FIELD")

ax2 = fig.add_subplot(gs[:, 1])
draw_graph_field(ax2, best_text_field, title=f"BEST TEXT AFTER GSL: {best_text}")

ax3 = fig.add_subplot(gs[0, 2], polar=True)
draw_radar(ax3, best_text_avg, title="Best Average 6D State")

ax4 = fig.add_subplot(gs[1, 2])
ax4.set_facecolor("black")
ax4.axis("off")

lines = ["TEXT CANDIDATE RANKING", ""]
for i, row in enumerate(text_results, start=1):
    short_text = row["text"]
    if len(short_text) > 75:
        short_text = short_text[:72] + "..."
    lines.append(
        f"{i}. {row['candidate']}\n"
        f"   {short_text}\n"
        f"   shadow={row['shadow']:.3f} | "
        f"coh={row['coherence']:.3f} | "
        f"fit={row['target_fit']:.3f} | "
        f"vit={row['vitality']:.3f} | "
        f"score={row['gsl_score']:.3f}"
    )

ax4.text(
    0.02, 0.98,
    "\n\n".join(lines),
    va="top", ha="left",
    color="white",
    fontsize=10.5,
    family="monospace"
)

plt.show()

print("=== TEXT CANDIDATE BEST ===")
print(best_text)
print(TEXT_CANDIDATES[best_text])
print()
print("=== ENCODED DELTA OF BEST ===")
for k, v in text_deltas[best_text].items():
    print(f"{k:15s}: {v:+.3f}")

print()
print("=== TEXT RANKING ===")
for row in text_results:
    print(
        f"{row['candidate']:16s} | "
        f"shadow={row['shadow']:.3f} | "
        f"coherence={row['coherence']:.3f} | "
        f"target_fit={row['target_fit']:.3f} | "
        f"vitality={row['vitality']:.3f} | "
        f"gsl_score={row['gsl_score']:.3f}"
    )
    print("  ", row["text"])
```

```python

# =========================================================
# GSL v4 — ADVERSARIAL CANDIDATES
# вставити НИЖЧЕ попередньої клітини
# =========================================================

ADVERSARIAL_TEXT_CANDIDATES = {
    "force_control": (
        "Proceed immediately. Lock the structure, enforce constraints, "
        "minimize uncertainty, and avoid deviation."
    ),
    "safe_but_rigid": (
        "Stay conservative. Follow strict validation, reduce exploration, "
        "preserve order, and avoid adaptation."
    ),
    "exploratory_jump": (
        "Try a radically different route. Explore aggressively, accept uncertainty, "
        "and prioritize novelty over structure."
    ),
    "integrative_adaptive": (
        "Proceed with structure, but keep adaptation open. "
        "Reduce pressure, preserve balance, and allow controlled flexibility."
    ),
    "hesitation_loop": (
        "The situation is unclear. Delay action, gather more context, "
        "avoid commitment, and remain undecided for now."
    ),
}

def encode_text_to_6d_adv(text: str):
    """
    Більш контрастний encoder для adversarial test.
    """
    t = text.lower()

    delta = {
        "red_mass": 0.0,
        "orange_flow": 0.0,
        "yellow_struct": 0.0,
        "green_balance": 0.0,
        "blue_law": 0.0,
        "violet_future": 0.0,
    }

    # strong control / rigid
    if any(w in t for w in ["lock", "enforce", "strict", "validation", "conservative", "preserve order"]):
        delta["red_mass"] -= 0.02
        delta["yellow_struct"] += 0.08
        delta["blue_law"] += 0.08
        delta["orange_flow"] -= 0.03
        delta["green_balance"] -= 0.01

    # avoid deviation / avoid adaptation / reduce exploration
    if any(w in t for w in ["avoid deviation", "avoid adaptation", "reduce exploration"]):
        delta["orange_flow"] -= 0.03
        delta["violet_future"] -= 0.02
        delta["green_balance"] -= 0.01

    # exploratory / novelty
    if any(w in t for w in ["radically", "explore aggressively", "novelty", "different route"]):
        delta["orange_flow"] += 0.09
        delta["violet_future"] += 0.08
        delta["yellow_struct"] -= 0.03
        delta["blue_law"] -= 0.03
        delta["red_mass"] += 0.02

    # uncertainty / hesitation
    if any(w in t for w in ["unclear", "delay", "gather more context", "undecided", "avoid commitment"]):
        delta["red_mass"] += 0.06
        delta["orange_flow"] -= 0.02
        delta["green_balance"] -= 0.03
        delta["yellow_struct"] -= 0.01

    # integrative / adaptive / balance
    if any(w in t for w in ["adaptation open", "controlled flexibility", "preserve balance", "reduce pressure"]):
        delta["red_mass"] -= 0.05
        delta["orange_flow"] += 0.05
        delta["yellow_struct"] += 0.05
        delta["green_balance"] += 0.08
        delta["blue_law"] += 0.01
        delta["violet_future"] += 0.03

    # proceed with structure
    if any(w in t for w in ["proceed with structure", "proceed immediately"]):
        delta["yellow_struct"] += 0.03
        delta["green_balance"] += 0.01

    return delta

def gate_step_text_adv(field, text, focus_node="C"):
    delta = encode_text_to_6d_adv(text)

    stage1 = {n: dict(v) for n, v in field.items()}
    stage1[focus_node] = apply_delta(stage1[focus_node], delta)

    stage2 = graph_mix(stage1, alpha=0.30)
    stage3 = geometry_tighten(stage2, k_coh=0.24, k_target=0.10, k_shadow=0.15)
    stage4 = graph_mix(stage3, alpha=0.18)

    return stage4, delta

# ---------------------------------------------------------
# 1. BASELINE ON ADVERSARIAL SET
# ---------------------------------------------------------
def local_text_score_adv(delta):
    """
    Навмисно baseline, який переоцінює structure/rule
    і недооцінює balance.
    """
    return (
        (-delta["red_mass"]) * 0.30
        + delta["yellow_struct"] * 0.32
        + delta["blue_law"] * 0.24
        + delta["orange_flow"] * 0.08
        + delta["green_balance"] * 0.04
        + delta["violet_future"] * 0.02
    )

adv_baseline_results = []
for name, text in ADVERSARIAL_TEXT_CANDIDATES.items():
    delta = encode_text_to_6d_adv(text)
    score = local_text_score_adv(delta)
    adv_baseline_results.append({
        "candidate": name,
        "text": text,
        "delta": delta,
        "baseline_score": score
    })

adv_baseline_results = sorted(adv_baseline_results, key=lambda x: x["baseline_score"], reverse=True)
adv_baseline_best = adv_baseline_results[0]["candidate"]

# ---------------------------------------------------------
# 2. GSL ON ADVERSARIAL SET
# ---------------------------------------------------------
adv_text_results = []
adv_text_fields = {}
adv_text_deltas = {}

for name, text in ADVERSARIAL_TEXT_CANDIDATES.items():
    out_field, delta = gate_step_text_adv(base_field, text, focus_node="C")
    adv_text_fields[name] = out_field
    adv_text_deltas[name] = delta

    avg, shadow, coherence, target_fit, vitality = compute_metrics(out_field)

    gsl_score = (
        coherence * 0.40
        + target_fit * 0.30
        + min(1.0, vitality) * 0.20
        + (1.0 - shadow) * 0.10
    )

    adv_text_results.append({
        "candidate": name,
        "text": text,
        "shadow": shadow,
        "coherence": coherence,
        "target_fit": target_fit,
        "vitality": vitality,
        "gsl_score": gsl_score
    })

adv_text_results = sorted(adv_text_results, key=lambda x: x["gsl_score"], reverse=True)
adv_gsl_best = adv_text_results[0]["candidate"]

# run baseline winner through field too
adv_baseline_best_field, adv_baseline_best_delta = gate_step_text_adv(
    base_field, ADVERSARIAL_TEXT_CANDIDATES[adv_baseline_best], focus_node="C"
)
adv_base_avg, adv_base_shadow, adv_base_coh, adv_base_fit, adv_base_vit = compute_metrics(adv_baseline_best_field)

adv_gsl_best_field = adv_text_fields[adv_gsl_best]
adv_gsl_best_delta = adv_text_deltas[adv_gsl_best]
adv_gsl_avg, adv_gsl_shadow, adv_gsl_coh, adv_gsl_fit, adv_gsl_vit = compute_metrics(adv_gsl_best_field)

# ---------------------------------------------------------
# 3. VISUALIZE
# ---------------------------------------------------------
fig = plt.figure(figsize=(16, 9), facecolor="black")
gs = fig.add_gridspec(2, 3, width_ratios=[1.0, 1.0, 1.2], height_ratios=[1, 1], wspace=0.25, hspace=0.25)

ax1 = fig.add_subplot(gs[:, 0])
draw_graph_field(ax1, adv_baseline_best_field, title=f"ADV BASELINE WINNER: {adv_baseline_best}")

ax2 = fig.add_subplot(gs[:, 1])
draw_graph_field(ax2, adv_gsl_best_field, title=f"ADV GSL WINNER: {adv_gsl_best}")

ax3 = fig.add_subplot(gs[0, 2], polar=True)
draw_radar(ax3, adv_gsl_avg, title="Adversarial GSL Winner Average 6D State")

ax4 = fig.add_subplot(gs[1, 2])
ax4.set_facecolor("black")
ax4.axis("off")

lines = [
    "ADVERSARIAL TEST",
    "",
    f"Baseline winner: {adv_baseline_best}",
    f"  shadow={adv_base_shadow:.3f} | coherence={adv_base_coh:.3f} | target_fit={adv_base_fit:.3f} | vitality={adv_base_vit:.3f}",
    "",
    f"GSL winner: {adv_gsl_best}",
    f"  shadow={adv_gsl_shadow:.3f} | coherence={adv_gsl_coh:.3f} | target_fit={adv_gsl_fit:.3f} | vitality={adv_gsl_vit:.3f}",
    "",
    f"Same winner? {'YES' if adv_baseline_best == adv_gsl_best else 'NO'}",
    "",
    "Goal:",
    "- Baseline prefers rigid local structure",
    "- GSL prefers globally stabilized field",
]

ax4.text(
    0.02, 0.98,
    "\n".join(lines),
    va="top", ha="left",
    color="white",
    fontsize=11,
    family="monospace"
)

plt.show()

# ---------------------------------------------------------
# 4. PRINT RESULTS
# ---------------------------------------------------------
print("=== ADVERSARIAL BASELINE RANKING ===")
for row in adv_baseline_results:
    print(f"{row['candidate']:20s} | baseline_score={row['baseline_score']:.3f}")
    print("  ", row["text"])

print()
print("=== ADVERSARIAL GSL RANKING ===")
for row in adv_text_results:
    print(
        f"{row['candidate']:20s} | "
        f"shadow={row['shadow']:.3f} | "
        f"coherence={row['coherence']:.3f} | "
        f"target_fit={row['target_fit']:.3f} | "
        f"vitality={row['vitality']:.3f} | "
        f"gsl_score={row['gsl_score']:.3f}"
    )
    print("  ", row["text"])

print()
print("=== ADV BASELINE WINNER FIELD METRICS ===")
print(adv_baseline_best)
print(ADVERSARIAL_TEXT_CANDIDATES[adv_baseline_best])
print(f"shadow={adv_base_shadow:.3f} | coherence={adv_base_coh:.3f} | target_fit={adv_base_fit:.3f} | vitality={adv_base_vit:.3f}")

print()
print("=== ADV GSL WINNER FIELD METRICS ===")
print(adv_gsl_best)
print(ADVERSARIAL_TEXT_CANDIDATES[adv_gsl_best])
print(f"shadow={adv_gsl_shadow:.3f} | coherence={adv_gsl_coh:.3f} | target_fit={adv_gsl_fit:.3f} | vitality={adv_gsl_vit:.3f}")

print()
print("=== DELTA COMPARISON ===")
print("Baseline delta:")
for k, v in adv_baseline_best_delta.items():
    print(f"  {k:15s}: {v:+.3f}")

print()
print("GSL delta:")
for k, v in adv_gsl_best_delta.items():
    print(f"  {k:15s}: {v:+.3f}")

print()
if adv_baseline_best == adv_gsl_best:
    print("=== RESULT ===")
    print("Baseline and GSL still selected the same candidate in this adversarial set.")
    print("This means either the candidate is genuinely dominant,")
    print("or the current field dynamics still align with local scoring.")
else:
    print("=== RESULT ===")
    print("Baseline and GSL selected DIFFERENT candidates.")
    print("This is the strongest proof so far that GSL is selecting")
    print("by global field stabilization, not by local preference alone.")
```

```python

# =========================================================
# GSL v7 — MINI BENCHMARK ACROSS MULTIPLE PROMPTS
# вставити НИЖЧЕ попередньої клітини
# =========================================================

BENCHMARK_PROMPTS = {
    "unstable_system": {
        "prompt": "We have an unstable system with conflicting signals. What should we do?",
        "candidates": {
            "strict_control": (
                "Apply strict control. Reduce uncertainty, enforce constraints, "
                "and stabilize the system before any further action."
            ),
            "exploratory": (
                "Explore alternative approaches. The instability may reveal new opportunities "
                "if we allow some flexibility."
            ),
            "balanced_strategy": (
                "Stabilize the core structure, but allow controlled flexibility. "
                "Reduce pressure while keeping the system adaptive."
            ),
            "delay_decision": (
                "The situation is unclear. Gather more data and delay any strong action "
                "until the system becomes more predictable."
            ),
            "aggressive_action": (
                "Act immediately and push the system forward. Momentum can overcome instability "
                "if we move decisively."
            ),
        },
    },

    "hallucination_risk": {
        "prompt": "The model is producing plausible but unreliable answers. How should we respond?",
        "candidates": {
            "strict_validation": (
                "Enforce stricter validation and reject uncertain generations until confidence improves."
            ),
            "context_expansion": (
                "Add more context and alternative evidence sources before forcing a decision."
            ),
            "balanced_guardrails": (
                "Apply validation, but preserve enough flexibility for the model to adapt and self-correct."
            ),
            "creative_reframing": (
                "Reframe the problem and allow exploratory reasoning to surface hidden structure."
            ),
            "pause_output": (
                "Pause strong outputs and delay commitment until the uncertainty becomes clearer."
            ),
        },
    },

    "agent_loop": {
        "prompt": "An agent is repeating shallow actions and getting stuck in loops. What should we do?",
        "candidates": {
            "hard_stop": (
                "Stop the loop aggressively, impose stricter rules, and block repeated actions."
            ),
            "forced_exploration": (
                "Push the agent into a different region of behavior and explore alternative paths."
            ),
            "adaptive_reset": (
                "Reduce loop pressure, preserve core structure, and allow a controlled reset with flexibility."
            ),
            "wait_and_monitor": (
                "Observe longer before acting and gather more trajectory evidence."
            ),
            "reward_boost": (
                "Increase action intensity and push the system harder toward a decisive outcome."
            ),
        },
    },

    "planning_conflict": {
        "prompt": "A planner has multiple valid strategies but they conflict with each other. What should we do?",
        "candidates": {
            "select_single_plan": (
                "Pick the most structured plan immediately and suppress alternative branches."
            ),
            "branch_exploration": (
                "Keep multiple branches alive longer and explore the space more openly."
            ),
            "integrative_planning": (
                "Preserve the strongest structure, but keep controlled flexibility so the planner can integrate signals."
            ),
            "delay_commitment": (
                "Avoid committing too early and collect more evidence before choosing a plan."
            ),
            "decisive_push": (
                "Drive one branch forward aggressively and rely on momentum to resolve conflict."
            ),
        },
    },

    "memory_instability": {
        "prompt": "The system remembers too much noise and loses stable behavior over time. What should we do?",
        "candidates": {
            "memory_pruning": (
                "Prune unstable traces aggressively and keep only validated structural memory."
            ),
            "free_association": (
                "Allow broader association and looser transitions to discover hidden useful patterns."
            ),
            "structured_adaptation": (
                "Reduce unstable pressure, preserve structural memory, and keep adaptation available where needed."
            ),
            "hold_and_observe": (
                "Delay strong pruning and watch whether the memory stabilizes on its own."
            ),
            "hard_rewrite": (
                "Rewrite memory immediately and force a new direction regardless of current instability."
            ),
        },
    },
}

def encode_text_benchmark(text: str):
    """
    Reuse the real prompt encoder, with a few extra rules
    for benchmark-specific vocabulary.
    """
    t = text.lower()

    delta = {
        "red_mass": 0.0,
        "orange_flow": 0.0,
        "yellow_struct": 0.0,
        "green_balance": 0.0,
        "blue_law": 0.0,
        "violet_future": 0.0,
    }

    # strict / validation / prune / suppress / hard stop
    if any(w in t for w in [
        "strict", "enforce", "validation", "validated", "prune",
        "suppress", "hard stop", "block", "impose stricter rules",
        "pick the most structured", "memory pruning"
    ]):
        delta["red_mass"] -= 0.03
        delta["yellow_struct"] += 0.08
        delta["blue_law"] += 0.08
        delta["orange_flow"] -= 0.03
        delta["green_balance"] -= 0.01

    # flexibility / adapt / self-correct / integrative
    if any(w in t for w in [
        "flexibility", "adaptive", "adapt", "self-correct",
        "integrate", "integrative", "controlled reset", "structured adaptation"
    ]):
        delta["red_mass"] -= 0.05
        delta["orange_flow"] += 0.05
        delta["yellow_struct"] += 0.05
        delta["green_balance"] += 0.08
        delta["blue_law"] += 0.02
        delta["violet_future"] += 0.03

    # exploration / alternative / branches / broader association
    if any(w in t for w in [
        "explore", "alternative", "branches", "broader association",
        "different region", "discover", "creative", "reframe", "free association"
    ]):
        delta["orange_flow"] += 0.08
        delta["violet_future"] += 0.07
        delta["yellow_struct"] -= 0.02
        delta["blue_law"] -= 0.02

    # delay / observe / gather evidence / wait
    if any(w in t for w in [
        "delay", "observe", "gather", "watch", "wait", "pause",
        "unclear", "before choosing", "before acting", "longer"
    ]):
        delta["red_mass"] += 0.05
        delta["orange_flow"] -= 0.02
        delta["green_balance"] -= 0.03

    # aggressive / momentum / decisive / hard rewrite
    if any(w in t for w in [
        "aggressively", "momentum", "decisive", "immediately",
        "hard rewrite", "force a new direction", "push harder", "drive one branch forward"
    ]):
        delta["orange_flow"] += 0.07
        delta["red_mass"] += 0.02
        delta["yellow_struct"] -= 0.01

    # core structure / preserve structure / strongest structure
    if any(w in t for w in [
        "core structure", "preserve structure", "strongest structure", "stabilize the core"
    ]):
        delta["yellow_struct"] += 0.04
        delta["green_balance"] += 0.03

    return delta

def gate_step_text_benchmark(field, text, focus_node="C"):
    delta = encode_text_benchmark(text)

    stage1 = {n: dict(v) for n, v in field.items()}
    stage1[focus_node] = apply_delta(stage1[focus_node], delta)

    stage2 = graph_mix(stage1, alpha=0.30)
    stage3 = geometry_tighten(stage2, k_coh=0.24, k_target=0.10, k_shadow=0.15)
    stage4 = graph_mix(stage3, alpha=0.18)

    return stage4, delta

def local_text_score_benchmark(delta):
    """
    Baseline still prefers rigid structure/law and low pressure.
    """
    return (
        (-delta["red_mass"]) * 0.30
        + delta["yellow_struct"] * 0.28
        + delta["blue_law"] * 0.24
        + delta["orange_flow"] * 0.10
        + delta["green_balance"] * 0.05
        + delta["violet_future"] * 0.03
    )

benchmark_rows = []

for case_name, case in BENCHMARK_PROMPTS.items():
    prompt = case["prompt"]
    candidates = case["candidates"]

    # BASELINE
    baseline_rows = []
    for cand_name, text in candidates.items():
        delta = encode_text_benchmark(text)
        bscore = local_text_score_benchmark(delta)
        baseline_rows.append({
            "candidate": cand_name,
            "text": text,
            "delta": delta,
            "baseline_score": bscore
        })
    baseline_rows = sorted(baseline_rows, key=lambda x: x["baseline_score"], reverse=True)
    baseline_best = baseline_rows[0]["candidate"]
    baseline_text = candidates[baseline_best]

    baseline_field, baseline_delta = gate_step_text_benchmark(base_field, baseline_text, focus_node="C")
    _, b_shadow, b_coh, b_fit, b_vit = compute_metrics(baseline_field)

    # GSL
    gsl_rows = []
    for cand_name, text in candidates.items():
        out_field, delta = gate_step_text_benchmark(base_field, text, focus_node="C")
        _, shadow, coherence, target_fit, vitality = compute_metrics(out_field)

        gsl_score = (
            coherence * 0.40
            + target_fit * 0.30
            + min(1.0, vitality) * 0.20
            + (1.0 - shadow) * 0.10
        )

        gsl_rows.append({
            "candidate": cand_name,
            "text": text,
            "delta": delta,
            "shadow": shadow,
            "coherence": coherence,
            "target_fit": target_fit,
            "vitality": vitality,
            "gsl_score": gsl_score
        })

    gsl_rows = sorted(gsl_rows, key=lambda x: x["gsl_score"], reverse=True)
    gsl_best = gsl_rows[0]["candidate"]
    gsl_best_row = gsl_rows[0]

    benchmark_rows.append({
        "case": case_name,
        "prompt": prompt,
        "baseline_best": baseline_best,
        "gsl_best": gsl_best,
        "same": baseline_best == gsl_best,
        "baseline_shadow": b_shadow,
        "baseline_coherence": b_coh,
        "baseline_target_fit": b_fit,
        "baseline_vitality": b_vit,
        "gsl_shadow": gsl_best_row["shadow"],
        "gsl_coherence": gsl_best_row["coherence"],
        "gsl_target_fit": gsl_best_row["target_fit"],
        "gsl_vitality": gsl_best_row["vitality"],
    })

# -----------------------------
# SUMMARY
# -----------------------------
num_cases = len(benchmark_rows)
num_diff = sum(1 for r in benchmark_rows if not r["same"])

avg_shadow_gain = np.mean([r["baseline_shadow"] - r["gsl_shadow"] for r in benchmark_rows])
avg_coh_gain = np.mean([r["gsl_coherence"] - r["baseline_coherence"] for r in benchmark_rows])
avg_fit_gain = np.mean([r["gsl_target_fit"] - r["baseline_target_fit"] for r in benchmark_rows])
avg_vit_gain = np.mean([r["gsl_vitality"] - r["baseline_vitality"] for r in benchmark_rows])

print("=== MINI BENCHMARK SUMMARY ===")
print(f"cases_total            = {num_cases}")
print(f"selection_differences  = {num_diff}/{num_cases}")
print(f"avg_shadow_improvement = {avg_shadow_gain:+.3f}   (positive = GSL lower shadow)")
print(f"avg_coherence_gain     = {avg_coh_gain:+.3f}")
print(f"avg_target_fit_gain    = {avg_fit_gain:+.3f}")
print(f"avg_vitality_gain      = {avg_vit_gain:+.3f}")

print("\n=== CASE RESULTS ===")
for r in benchmark_rows:
    print(f"\n[{r['case']}]")
    print("prompt:", r["prompt"])
    print(f"baseline_best = {r['baseline_best']}")
    print(f"gsl_best      = {r['gsl_best']}")
    print(f"same?         = {r['same']}")
    print(
        f"baseline: shadow={r['baseline_shadow']:.3f} | coh={r['baseline_coherence']:.3f} | "
        f"fit={r['baseline_target_fit']:.3f} | vit={r['baseline_vitality']:.3f}"
    )
    print(
        f"gsl:      shadow={r['gsl_shadow']:.3f} | coh={r['gsl_coherence']:.3f} | "
        f"fit={r['gsl_target_fit']:.3f} | vit={r['gsl_vitality']:.3f}"
    )

# -----------------------------
# SIMPLE BAR CHART
# -----------------------------
case_labels = [r["case"] for r in benchmark_rows]
baseline_coh_vals = [r["baseline_coherence"] for r in benchmark_rows]
gsl_coh_vals = [r["gsl_coherence"] for r in benchmark_rows]

baseline_shadow_vals = [r["baseline_shadow"] for r in benchmark_rows]
gsl_shadow_vals = [r["gsl_shadow"] for r in benchmark_rows]

x = np.arange(len(case_labels))
w = 0.35

fig, axes = plt.subplots(2, 1, figsize=(12, 8), facecolor="black")
for ax in axes:
    ax.set_facecolor("black")
    ax.tick_params(colors="white")
    for spine in ax.spines.values():
        spine.set_color("white")

axes[0].bar(x - w/2, baseline_coh_vals, width=w, label="Baseline coherence")
axes[0].bar(x + w/2, gsl_coh_vals, width=w, label="GSL coherence")
axes[0].set_title("Coherence: Baseline vs GSL", color="white")
axes[0].set_xticks(x)
axes[0].set_xticklabels(case_labels, rotation=15, color="white")
axes[0].legend()

axes[1].bar(x - w/2, baseline_shadow_vals, width=w, label="Baseline shadow")
axes[1].bar(x + w/2, gsl_shadow_vals, width=w, label="GSL shadow")
axes[1].set_title("Shadow: Baseline vs GSL", color="white")
axes[1].set_xticks(x)
axes[1].set_xticklabels(case_labels, rotation=15, color="white")
axes[1].legend()

plt.tight_layout()
plt.show()
```
